```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)

# ============================================================================
# LOAD DATA
# ============================================================================
# Update this path to match your file location


print("="*80)
print("XCS MATCH SET ANALYSIS - Feature Importance for Fertilizer Decision")
print("="*80)
print(f"\nTotal classifiers: {len(df)}")
print(f"Unique actions: {sorted(df['action'].unique())}")
print(f"\nAction distribution:")
print(df['action'].value_counts().sort_index())

# ============================================================================
# 1. IDENTIFY FEATURES (State Variables)
# ============================================================================
# Features are the state variables with _lo, _hi, _mask suffixes
feature_names = [
    'DVS',           # Development Stage
    'LAI',           # Leaf Area Index
    'RFTRA',         # Radiation Use Efficiency
    'WSO',           # Weight Storage Organs
    'Nsurp_sofar',   # Nitrogen surplus so far
    'last_dose_kg_ha',      # Last fertilizer dose
    'last_app_gap_days',    # Days since last application
    'cum_N_applied_kg_ha',  # Cumulative N applied
    'IRRAD_7',       # 7-day radiation
    'TMEAN_7',       # 7-day mean temperature
    'RAIN_7',        # 7-day rainfall
    'Ndep_sum',      # Nitrogen deposition sum
    'SM_mean'        # Soil moisture mean
]

print("\n" + "="*80)
print("FEATURES ANALYZED:")
print("="*80)
for i, feat in enumerate(feature_names, 1):
    print(f"{i:2d}. {feat}")

# ============================================================================
# 2. FEATURE SPECIFICITY ANALYSIS
# ============================================================================
# Features with mask=0 are specific (used in the condition)
# Features with mask=1 are don't-care (not used)

def calculate_feature_specificity(df, feature_names):
    """Calculate how often each feature is specified (not don't-care)"""
    specificity = {}

    for feat in feature_names:
        mask_col = f"{feat}_mask"
        if mask_col in df.columns:
            # mask=0 means specific, mask=1 means don't-care
            specificity[feat] = (df[mask_col] == 0).sum() / len(df) * 100

    return pd.Series(specificity).sort_values(ascending=False)

specificity = calculate_feature_specificity(df, feature_names)

print("\n" + "="*80)
print("FEATURE SPECIFICITY (% of classifiers using each feature)")
print("="*80)
print(specificity)

# ============================================================================
# 3. FEATURE RANGE ANALYSIS BY ACTION
# ============================================================================
def analyze_feature_ranges_by_action(df, feature_names):
    """Analyze the ranges of features for different actions"""

    results = []

    for feat in feature_names:
        lo_col = f"{feat}_lo"
        hi_col = f"{feat}_hi"
        mask_col = f"{feat}_mask"

        if lo_col not in df.columns or hi_col not in df.columns:
            continue

        # Only look at specific (non-don't-care) conditions
        df_specific = df[df[mask_col] == 0]

        if len(df_specific) == 0:
            continue

        for action in sorted(df['action'].unique()):
            df_action = df_specific[df_specific['action'] == action]

            if len(df_action) > 0:
                results.append({
                    'Feature': feat,
                    'Action': action,
                    'Count': len(df_action),
                    'Mean_lo': df_action[lo_col].mean(),
                    'Mean_hi': df_action[hi_col].mean(),
                    'Range_width': (df_action[hi_col] - df_action[lo_col]).mean()
                })

    return pd.DataFrame(results)

range_analysis = analyze_feature_ranges_by_action(df, feature_names)

print("\n" + "="*80)
print("FEATURE RANGES BY ACTION (for specified features only)")
print("="*80)
print(range_analysis.to_string())

# ============================================================================
# 4. WEIGHTED FEATURE IMPORTANCE (by Fitness)
# ============================================================================
def calculate_weighted_importance(df, feature_names):
    """Weight feature importance by classifier fitness"""

    importance = {}

    for feat in feature_names:
        mask_col = f"{feat}_mask"
        if mask_col in df.columns:
            # Sum of fitness for classifiers that specify this feature
            specific_fitness = df[df[mask_col] == 0]['F'].sum()
            total_fitness = df['F'].sum()

            if total_fitness > 0:
                importance[feat] = (specific_fitness / total_fitness) * 100

    return pd.Series(importance).sort_values(ascending=False)

weighted_importance = calculate_weighted_importance(df, feature_names)

print("\n" + "="*80)
print("WEIGHTED FEATURE IMPORTANCE (by classifier fitness)")
print("="*80)
print(weighted_importance)

# ============================================================================
# 5. CORRELATION WITH PREDICTION VALUES
# ============================================================================
def correlation_with_prediction(df, feature_names):
    """Analyze correlation between feature values and predictions"""

    correlations = {}

    for feat in feature_names:
        lo_col = f"{feat}_lo"
        hi_col = f"{feat}_hi"
        mask_col = f"{feat}_mask"

        if lo_col not in df.columns or mask_col not in df.columns:
            continue

        # Only look at specific conditions
        df_specific = df[df[mask_col] == 0]

        if len(df_specific) > 1:
            # Use midpoint of range
            midpoint = (df_specific[lo_col] + df_specific[hi_col]) / 2
            corr = midpoint.corr(df_specific['p'])
            correlations[feat] = corr

    return pd.Series(correlations).sort_values(ascending=False)

correlations = correlation_with_prediction(df, feature_names)

print("\n" + "="*80)
print("CORRELATION WITH PREDICTION VALUE (p)")
print("="*80)
print(correlations)

# ============================================================================
# 6. VISUALIZATION: Feature Specificity
# ============================================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Feature Specificity
ax = axes[0, 0]
specificity.plot(kind='barh', ax=ax, color='steelblue')
ax.set_xlabel('% of Classifiers Using Feature')
ax.set_title('Feature Specificity\n(Higher = More Important)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)

# Plot 2: Weighted Importance
ax = axes[0, 1]
weighted_importance.plot(kind='barh', ax=ax, color='coral')
ax.set_xlabel('Fitness-Weighted Importance (%)')
ax.set_title('Weighted Feature Importance\n(by Classifier Fitness)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)

# Plot 3: Action Distribution
ax = axes[1, 0]
action_counts = df['action'].value_counts().sort_index()
action_counts.plot(kind='bar', ax=ax, color='mediumseagreen')
ax.set_xlabel('Action (Fertilizer Amount)')
ax.set_ylabel('Number of Classifiers')
ax.set_title('Action Distribution in Match Set', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Plot 4: Correlation with Prediction
ax = axes[1, 1]
correlations.plot(kind='barh', ax=ax, color='mediumpurple')
ax.set_xlabel('Correlation with Prediction (p)')
ax.set_title('Feature Correlation with Predicted Reward', fontsize=12, fontweight='bold')
ax.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('xcs_feature_importance.png', dpi=300, bbox_inches='tight')
print("\nVisualization saved as 'xcs_feature_importance.png'")

# ============================================================================
# 7. TOP FEATURES SUMMARY
# ============================================================================
print("\n" + "="*80)
print("TOP 5 MOST IMPORTANT FEATURES (Summary)")
print("="*80)

# Combine metrics
top_features = pd.DataFrame({
    'Specificity (%)': specificity,
    'Weighted Importance (%)': weighted_importance,
    'Correlation with p': correlations
})

top_features['Combined_Score'] = (
    top_features['Specificity (%)'] / top_features['Specificity (%)'].max() +
    top_features['Weighted Importance (%)'] / top_features['Weighted Importance (%)'].max() +
    abs(top_features['Correlation with p']) / abs(top_features['Correlation with p']).max()
) / 3

top_features = top_features.sort_values('Combined_Score', ascending=False)

print(top_features.head(10).to_string())

# ============================================================================
# 8. DETAILED ANALYSIS FOR TOP FEATURES
# ============================================================================
print("\n" + "="*80)
print("DETAILED ANALYSIS FOR TOP 3 FEATURES")
print("="*80)

top_3 = top_features.head(3).index.tolist()

for feat in top_3:
    print(f"\n{feat}:")
    print("-" * 40)

    lo_col = f"{feat}_lo"
    hi_col = f"{feat}_hi"
    mask_col = f"{feat}_mask"

    df_specific = df[df[mask_col] == 0]

    if len(df_specific) > 0:
        for action in sorted(df_specific['action'].unique()):
            df_action = df_specific[df_specific['action'] == action]
            if len(df_action) > 0:
                print(f"  Action {action}: {len(df_action)} classifiers")
                print(f"    Range: [{df_action[lo_col].mean():.3f}, {df_action[hi_col].mean():.3f}]")
                print(f"    Avg Fitness: {df_action['F'].mean():.3f}")

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
```
